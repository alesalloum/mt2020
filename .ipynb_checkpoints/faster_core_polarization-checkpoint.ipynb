```python
#polarized_net = gt.load_graph("data/networks/jurassicworld.graphml")
polarized = gt.Graph(gt.collection.data["polblogs"], directed=False)

l = gt.topology.label_largest_component(polarized)
polarized_net = gt.GraphView(polarized, vfilt=l)
gt.draw.graph_draw(polarized_net)
```

```python
## Find the cores with graph-tool

# CORE I
kcore = gt.topology.kcore_decomposition(polarized_net)
polarized_net.vertex_properties["kcore"] = kcore

gt.draw.graph_draw(polarized_net, vertex_fill_color=kcore, vertex_text=kcore)

largest_core = gt.util.find_vertex(polarized_net, kcore, max(kcore))
vfilt = polarized_net.new_vertex_property('bool')

for node in largest_core:
    vfilt[node] = True
    
polarized_net.vertex_properties["lc"] = vfilt

# Visualize the first core
gt.draw.graph_draw(polarized_net, vertex_fill_color=vfilt, vertex_text=kcore)

# CORE II
polarized_net.set_vertex_filter(vfilt, inverted=True)
gt.draw.graph_draw(polarized_net)

l = gt.topology.label_largest_component(polarized_net)
u = gt.GraphView(polarized_net, vfilt=l)
gt.draw.graph_draw(u)

kcore = gt.topology.kcore_decomposition(u)
u.vertex_properties["kcore"] = kcore

largest_core = gt.util.find_vertex(u, kcore, max(kcore))

vfilt = u.new_vertex_property('bool')

for node in largest_core:
    vfilt[node] = True
    
u.vertex_properties["lc"] = vfilt

gt.draw.graph_draw(u, vertex_fill_color=vfilt, vertex_text=kcore)
```

```python
def complete(polarized_net):
    
    ## Find the cores with graph-tool
    # CORE I
    kcore = gt.topology.kcore_decomposition(polarized_net)
    polarized_net.vertex_properties["kcore"] = kcore

    gt.draw.graph_draw(polarized_net, vertex_fill_color=kcore, vertex_text=kcore)

    largest_core = gt.util.find_vertex(polarized_net, kcore, max(kcore))

    vfilt = polarized_net.new_vertex_property('bool')

    for node in largest_core:
        vfilt[node] = True

    polarized_net.vertex_properties["lc1"] = vfilt

    # Visualize the first core
    gt.draw.graph_draw(polarized_net, vertex_fill_color=vfilt, vertex_text=kcore)

    # CORE II
    inter_net = gt.Graph(polarized_net, directed=False)
    
    inter_net.set_vertex_filter(vfilt, inverted=True)
    gt.draw.graph_draw(inter_net)

    l = gt.topology.label_largest_component(inter_net)
    u = gt.GraphView(inter_net, vfilt=l)
    gt.draw.graph_draw(u)

    kcore = gt.topology.kcore_decomposition(u)
    u.vertex_properties["kcore"] = kcore

    largest_core = gt.util.find_vertex(u, kcore, max(kcore))

    vfilt = polarized_net.new_vertex_property('bool')
    vfiltu = polarized_net.new_vertex_property('bool')

    for node in largest_core:
        vfilt[node] = True
        vfiltu[node] = True

    u.vertex_properties["lc2"] = vfilt
    polarized_net.vertex_properties["lc2"] = vfilt
    gt.draw.graph_draw(u, vertex_fill_color=vfilt, vertex_text=kcore)
    
    return polarized_net
```

```python
polarized_net = complete(polarized_net)
```
